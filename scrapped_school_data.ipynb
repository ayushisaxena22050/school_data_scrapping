```python
#importing necessary libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import requests
# import lxml.html as lh
import pandas as pd
from bs4 import BeautifulSoup
import time
import numpy as np
import csv
columns=['School'
,'Instruction_Medium',
 'Meal_Not_Applicable',
 'Male_Teachers',
 'Pre_Primary_Sectin_Avilable',
 'Board_for_Class_10th_N/A',
 'School_Type',
 'Classes',
 'Female_Teacher',
 'Pre_Primary_Teachers',
 'Board_for_Class_10+2',
 'Establishment',
 'School_Area',
 'Management',
 'School_Shifted_to_New_Place',
 'Head_Teachers',
 'Head_Teacher',
 'Is_School_Residential',
 'Residential_Type',
 'Total Teachers',
 'Contract Teachers',
 'Pincode',
 'Village/Town',
 'Cluster',
 'Block',
 'District',
 'State',
 'UDISE_Code',
 'Building',
 'Class_Rooms',
 'Boys_Toilet',
 'Girls_Toilet',
 'Computer_Aided_Learning',
 'Electricity',
 'Wall',
 'Library',
 'Playground',
 'Books_in_Library',
 'Drinking_Water',
 'Ramps_for_Disable',
 'Computers']
```

```python
url='https://schools.org.in/'
timeout=200
browser=webdriver.Chrome(executable_path=r'C:\Users\181635\Downloads\chromedriver.exe')
browser.get(url)
browser.set_page_load_timeout(timeout)
browser.implicitly_wait(30)
elem=browser.find_element_by_link_text('Schools in Punjab')
elem.click()
districts=browser.find_elements_by_xpath('//table[@class="table_1"]//td[2]')
browser.implicitly_wait(10)
districts=[district.text for district in districts]

for district in range(13,14):
    print(districts[district])
    browser.implicitly_wait(30)
    elem2=browser.find_element_by_link_text(districts[district])
    elem2.click()
    blocks=browser.find_elements_by_xpath('//table[@class="table_1"]//td[2]')
    browser.implicitly_wait(30)
    blocks=[block.text for block in blocks]
    
    for block in range(len(blocks)):
        print(blocks[block])
        browser.implicitly_wait(30)
        elem3=browser.find_element_by_link_text(blocks[block])
        elem3.click()
        clusters=browser.find_elements_by_xpath('//table[@class="table_1"]//td[2]')
        browser.implicitly_wait(30)
        clusters=[cluster.text for cluster in clusters]
#         print(clusters)
        for cluster in range(len(clusters)):
            print('========',clusters[cluster])
            browser.implicitly_wait(30)
            elem4=browser.find_element_by_link_text(clusters[cluster])
            elem4.click()
            schools=browser.find_elements_by_xpath('//table[@class="table_1"]//td[2]')
            browser.implicitly_wait(30)
            schools=[school.text for school in schools]
            if schools:
                for school in range(0,len(schools)):
                    data=[]
                    info=[]
                    data.append(schools[school])
                    browser.implicitly_wait(30)
                    elem5=browser.find_element_by_link_text(schools[school])
                    elem5.click()
                    html=BeautifulSoup(browser.page_source)
                    WebDriverWait(browser, 20).until(EC.presence_of_all_elements_located((By.CLASS_NAME,'content-boxed')))
                    class1=html.find(class_='content-boxed')
                    tags=class1.findAll('b')
                    browser.implicitly_wait(30)
                    if tags:
                        for tag in tags:
                            data.append(tag.text)
                        WebDriverWait(browser, 20).until(EC.presence_of_all_elements_located((By.CLASS_NAME,'listings-price-widget')))
                        class2=html.find(class_="listings-price-widget")
                        browser.implicitly_wait(30)
                        pincode=class2.text.split(":")[1]
                        data.append(pincode)
                        WebDriverWait(browser, 20).until(EC.presence_of_all_elements_located((By.ID,'house-details-sidebar')))
                        class3=html.find(id="house-details-sidebar")
                        browser.implicitly_wait(30)
                        names=class3.findAll('span')
                        for cl in names:
                            data.append(cl.text)
#                     else:
#                         data=[0]*41
                    info.append(data)
                    with open("Punjab_data.csv", "a", newline="") as f:
                        writer = csv.writer(f,columns)
                        writer.writerows(info)
                    f.close()
                    browser.execute_script("window.history.go(-1)")
            time.sleep(2)
            browser.back()
        time.sleep(5)
        browser.back()
    time.sleep(7)
    browser.back()
browser.quit()
```
